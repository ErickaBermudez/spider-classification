# Spider Recognition with CNN 1. Data Exploration

In this notebook we will explore our data. 

```python
# imports
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt
import matplotlib as mpl
import seaborn as sns
import os
```

```python
# first let's check our csv to check what kind of data we have and get our labels
spiders_df = pd.read_csv('data/spiders.csv')
spiders_df.head()
```

We can see 3 columns: the index of the spider, its filepath and it label. Let's check our spiders.

```python
print("--------Spiders in our dataset----------")
spiders_count = spiders_df['labels'].value_counts()
print("Total of spiders: ", len(spiders_count))
spiders_count
```

```python
# plotting to visualize better 
spiders_count.plot.bar()
plt.show()
```

```python
# it looks very nice! now we will check our images
spiders_list = os.listdir('data/train')
print("Our spider folders in train")
for spider_folder_name in spiders_list: 
    print("'"+spider_folder_name+"'")
```

```python
# we already know the structure of our filepath 
# train\Black Widow\001.jpg
# so we can get one of each set
plt.figure()
# making our figure bigger
plt.figure(figsize=(20, 10))

for index, spider_name in enumerate(spiders_list):
    # we can get the n image from all the spiders
    spider_path = 'data/train/' + spider_name + "/001.jpg"
    spider_img = plt.imread(spider_path)
    plt.subplot(3, 5, index + 1)
    plt.imshow(spider_img)
    plt.axis('off')
    plt.title(spider_name)
plt.show()
```

# Conclusions of data exploration
This is a very clean dataset and it will not be hard to work with it. The spider pictures' quality we had the chance to see had from high to reasonable quality and are very diverse (as in, there are a lot of pictures for each spider and from different angles), which will be good for our model but maybe not that good if we later try to implement an application in real life. 

Another disadvantage is that is a pretty limited dataset. There are over 48,000 recognized species of spiders worldwide [1] 15 spiders are not too many spiders and our results might not be impactful.
